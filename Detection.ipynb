# Exploration Code

```python
import tensorflow as tf
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2, preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt

# Load pre-trained MobileNetV2 model
model = MobileNetV2(weights='imagenet')

# Load and preprocess the image
img_path = "C:\Users\Admin\Pictures\Screenshots\Screenshot (100).png"  # Replace with the path to your image
img = image.load_img(img_path, target_size=(224, 224))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array = preprocess_input(img_array)

# Predict the class of the image
predictions = model.predict(img_array)

# Decode the predictions into class names
decoded_predictions = decode_predictions(predictions, top=3)[0]

# Display the top 3 predictions with their scores
for i, (imagenet_id, label, score) in enumerate(decoded_predictions):
    print(f"{i + 1}. {label}: {score * 100:.2f}%")
```

# Functional Code

```python


def predict_image_class(img_path):
    # Load and preprocess the image
    img = image.load_img(img_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = preprocess_input(img_array)

    # Predict the class of the image
    predictions = model.predict(img_array)

    # Decode the predictions into class names
    decoded_predictions = decode_predictions(predictions, top=5)[0]

    # Display the image
    plt.imshow(img)
    plt.axis('off')  # Hide axis
    plt.show()

    
    # Display the top 3 predictions with their scores
    for i, (imagenet_id, label, score) in enumerate(decoded_predictions):
        print(f"{i + 1}. {label}: {score * 100:.2f}%")
    

# Test the function with animal, vehicle, and human images
img_path = "C:\Users\Admin\Pictures\Screenshots\Screenshot (100).png"
predict_image_class(img_path)
```
